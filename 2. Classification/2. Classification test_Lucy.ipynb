# Purpose
研究婚外情的數據

(1) 該位外遇的可能性有多大？ 
(2) 影響外遇的原因是什麼？ 
(3) 可以根據分析結果說明如何降低外遇的發生嗎？

# Answer


### (1)
因為樣本只有601位，且各項變量類別和數值分佈較極端，要以此數據預測一個人外遇的可能性，精確度過低，從模型表現看起來也是如此。
### (2)
以目前樣本來看，影響外遇的因素主要包含yearsmarried（婚姻時間）、rating（婚姻評分），普遍而言結婚越久容易外遇;另外有許多因素可能影響rating， 婚姻評分給1分和2分者，有小孩的佔比超過八成，教育程度高低跟職業別與是否外遇皆無明顯相關性。
### (3)
因為外遇是不可預測的，所以降低外遇發生這題本身就無解。若只以這個樣本的數據結果來說，降低外遇發生=讓婚姻評分保持高分。要維持婚姻評分，除了夫妻之間無法量化的情感指標外，外在條件上建議有穩定的工作收入，並維持宗教信仰，另外小孩可能也是影響評分的關鍵因素，不過從資料無法得知每位受測者生了幾個小孩，較難分析大概生幾個小孩會導致婚姻評分下降。

# 數據說明

* 婚外情數據即著名的“Fair’s Affairs”，取自於1969年《今日心理》（Psychology Today）所做的一個代表性調查
* 該數據從601個參與者身上收集了9個變量
* 變數包括一年來婚外情的頻率以及參與者性別、年齡、婚齡、是否有小孩、宗教信仰程度、學歷、職業，還有對婚姻的自我評分

#### affairs 受訪者在過去一年中進行外遇的頻率         
0 = 無         
1 = 一次       
2 = 兩次         
3 = 三倍         
7 = 4 - 10 次         
12 = 每月或更多         

#### gender          
0 = 女性         
1 = 男性         

#### age         
17.5 = 20 歲以下         
22.0 = 20 - 24         
27.0 = 25 - 29         
32.0 = 30 - 34         
37.0 = 35 - 39         
42.0 = 40 - 44         
47.0 = 45 - 49         
52.0 = 50 - 54         
57.0 = 55 或以上         

#### yearsmarried  婚姻時間                  
0.125 = 3 個月或更短         
0.417 = 4 - 6 個月         
0.750 = 6 個月 - 1 年         
1.500 = 1 - 2 年         
4.000 = 3 - 5 年         
7.000 = 6 - 8 年         
10.00 = 9 - 11 年         
15.00 = 12 年或更長時間         

#### children 孩子人數                  
0 = 無         
1 = 一個或多個         
         
#### religiousness 婚內的宗教信仰         
1 = 反宗教         
2 = 完全沒有         
3 = 輕微         
4 = 有點         
5 = 非常         

#### education 教育程度         
9.0 = 小學         
12.0 = 高中畢業         
14.0 = 一些大學         
16.0 = 大學畢業生         
17.0 = 一些畢業作品         
18.0 = 碩士學位         
20.0 = 博士、醫學博士或其他高級學位 

#### occupation 詳見data-descriptions.pdf

#### rating 對婚姻的自我評分 (5分制，1表示非常不幸福，5表示非常幸福）

# 建模

```python
#Convertin the predictor variable in a binary numeric variable
df['ynaffairs'].replace(to_replace=1, value=1, inplace=True)
df['ynaffairs'].replace(to_replace=0,  value=0, inplace=True)
df=df.drop(columns = ['affairs'])
#Let's convert all the categorical variables into dummy variables
df_dummies = pd.get_dummies(df)
df_dummies.head()
```

```python
df_dummies.columns
```

```python
# We will use the data frame where we had created dummy variables
X = df_dummies.drop(columns = ['ynaffairs'])
y = df_dummies['ynaffairs'].values

# Scaling all the variables to a range of 0 to 1
from sklearn.preprocessing import MinMaxScaler
features = X.columns.values
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(X)
X = pd.DataFrame(scaler.transform(X))
X.columns = features
```

```python
# Create Train & Test Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=777)
```

# Logistic Regression

```python
# Running logistic regression model
from sklearn.linear_model import LogisticRegression
model_log = LogisticRegression()
result = model_log.fit(X_train, y_train)
```
